```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("education_costs.csv")
print(df.head())
print(df.describe())
print(df.info())
```

```python
df["Total"] = df.Tuition_USD + df.Living_Cost_Index+ df.Rent_USD+ df.Visa_Fee_USD+ df.Insurance_USD
```

```python
df.head()
```

```python
# 1. Get the initial number of rows
initial_rows = len(df)
print(f"Initial number of rows: {initial_rows}")

# 2. Drop rows containing any null values
df_dropped = df.dropna()

# 3. Get the number of rows after dropping nulls
rows_after_dropna = len(df_dropped)
print(f"Number of rows after dropping NaN: {rows_after_dropna}")

# 4. Calculate the number of rows dropped
rows_dropped = initial_rows - rows_after_dropna
print(f"Number of rows dropped: {rows_dropped}")

# 5. Calculate the percentage of data erased
percentage_erased = (rows_dropped / initial_rows) * 100
print(f"Percentage of data erased: {percentage_erased:.2f}%")
```
